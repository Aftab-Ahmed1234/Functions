```python
def countdown(n):
 if n<=0:
    print("Done!")
 else:
    print(n)
    countdown(n-1)
countdown(5)
```

```python
amount=int(input("Enter amount  :"))
r1=amount%5000
fk=amount-r1

fk==5000
note=fk/5000;
print("5000 notes are",note)

r2=r1%1000
ok=r2-r1
ok==1000
note=ok/1000;
print(" 1000 notes are",note)

r3=r2%500
fh=r3-r2
fh==500
note=fh/500;
print("500 notes are",note)
    

r4=r3%100
oh=r4-r3
oh==100
note=oh/100;
print("100 notes are",note)

r5=r4%50
fif=r5-r4
fif==50
note=fif/50;
print("50 notes are",note)

r6=r5%20
twonty=r6-r5
twonty==20
note=twonty/20;
print("20 notes are",note)

r7=r6%10
ten=r7-r6
ten==10
note=ten/10;
print("10 notes are",note)
```
