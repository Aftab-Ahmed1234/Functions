```python
amount=int(input("Enter amount  :"))

r1=amount%5000
fk=amount-r1

fk==5000
note=fk/5000;
print("5000 notes are",int(note))

r2=r1%1000
ok=r1-r2
ok==1000
note=ok/1000;
print(" 1000 notes are",int(note))

r3=r2%500
fh=r2-r3
fh==500
note=fh/500;
print("500 notes are",int(note))
    

r4=r3%100
oh=r3-r4
oh==100
note=oh/100;
print("100 notes are",int(note))

r5=r4%50
fif=r4-r5
fif==50
note=fif/50;
print("50 notes are",int(note))

r6=r5%20
twenty=r5-r6
twenty==20
note=twenty/20;
print("20 notes are",int(note))

r7=r6%10
ten=r6-r7
ten==10
note=ten/10;
print("10 notes are",int(note))
```

```python
amount = int(input("Enter amount: "))

notes = [5000, 1000, 500, 100, 50, 20, 10]

for note in notes:
    count = amount // note
    amount = amount % note
    print(f"{note} notes are {count}")
```

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))
```

```python
def fibonacii(n):
  if n<=1:
    return n
  else:
    return fibonacii(n-1)+fibonacii(n-1)
print(fibonacii(7))
```
